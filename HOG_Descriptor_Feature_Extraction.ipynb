# **Implementation of the HOG Descriptor**  

First, we will load the libraries and the image for which we will calculate the HOG descriptor:  

```python
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt
import cv2 as cv
%matplotlib inline

img = cv.imread('puppy.png',0)
plt.imshow(img,cmap='gray')
print(img.shape)
```

We first need to resize this image to 64 x 128. Then, we will directly use the ***hog*** function from ***skimage.features***. The ***hog*** function will internally compute the result and return the feature matrix.

```python
resized_img=cv.resize(img,(64,128))
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8),
                    cells_per_block=(2, 2), visualize=True)

print(fd.shape)
plt.plot(fd)
```

As expected, we have 3,780 features for the image, which confirms the calculations we performed during the course. You can choose to modify the hyperparameter values, which will result in a feature matrix of different sizes.  

Finally, let's look at the HOG image:  

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8), sharex=True, sharey=True)

ax1.imshow(resized_img, cmap=plt.cm.gray)
ax1.set_title('Input image')


ax2.imshow(hog_image, cmap=plt.cm.gray)
ax2.set_title('Histogram of Oriented Gradients')

plt.show()
```

# **Implementation of the LBP Descriptor**  
Now, let's calculate the LBP descriptor.  

```python
from skimage.feature import local_binary_pattern
import numpy as np

img = cv.imread('/content/elvis.png',0)
plt.imshow(img,cmap='gray')
```

```python
LBP=local_binary_pattern(img,8,1)
print(LBP.shape)
plt.imshow(LBP,cmap='gray')
```

```python
def histogram(img,l):
  hist=np.zeros(l,dtype=int)
  for i in range(img.shape[0]):
    for j in range(img.shape[1]):
      hist[img[i,j]]+=1
  return hist
```

```python
h=histogram(img,256)
plt.plot(h)
plt.show()
```
